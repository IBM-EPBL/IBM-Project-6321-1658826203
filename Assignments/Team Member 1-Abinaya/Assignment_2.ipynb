# 6.FINDING OUTLIERS AND REPLACING

```python
sns.boxplot(x=df["Balance"])
```

```python
q2 = df.Balance.quantile(0.25)
q3 = df.Balance.quantile(0.75)
IQR = q3 - q2
upper_limit = q3+1.5*IQR
lower_limit = q2-1.5*IQR
print(lower_limit)
print(upper_limit)
df["Balance"].head
```

```python
df["Balance"] = np.where(df["Balance"]<lower_limit,6.52,df["Balance"])
sns.violinplot(x=df["Balance"])
```

```python
sns.boxplot(x=df["Balance"])
```
